```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")

# Show first 5 rows
data.head()
```

```python
print(data.columns.tolist())
```

```python
data.columns = ['Region', 'Date', 'Frequency', 'Unemployment Rate (%)',
                'Employed', 'Labour Participation Rate (%)',
                'Region_1', 'Longitude', 'Latitude']
```

```python
data = data[['Region', 'Date', 'Unemployment Rate (%)']]
```

```python
# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Unemployment Rate (%)', data=data)
plt.title('Unemployment Rate Over Time in India')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='Region', y='Unemployment Rate (%)', data=data)
plt.xticks(rotation=90)
plt.title('Unemployment Rate by Region in India')
plt.show()
```

```python
region_avg = data.groupby('Region')['Unemployment Rate (%)'].mean().reset_index()

plt.figure(figsize=(14, 6))
sns.barplot(x='Region', y='Unemployment Rate (%)', data=region_avg)
plt.xticks(rotation=90)
plt.title('Average Unemployment Rate by Region')
plt.show()
```
